## __*Motivation for Dataset Creation:*__

### Why was the dataset created? (e.g., were there specific tasks in mind, or a specific gap that needed to be filled?)
	VAERS, Vaccine Adverse Event Reporting System, was established to be a national early warning system to detect possible safety problems in U.S. licensed vaccines. The system accepts and analyzes reports of adverse events and possible side effects after vaccination has occurred. It was initially established in 1990 as a response to the National Childhood Vaccine Injury Act.  

### 	Objectives via VAERS website: 
•	Detect new, unusual, or rare vaccine adverse events;
•	Monitor increases in known adverse events;
•	Identify potential patient risk factors for particular types of adverse events;
•	Assess the safety of newly licensed vaccines;
•	Determine and address possible reporting clusters (e.g., suspected localized [temporally or geographically] or product-/batch-/lot-specific adverse event reporting);
•	Recognize persistent safe-use problems and administration errors;
•	Provide a national safety monitoring system that extends to the entire general population for response to public health emergencies, such as a large-scale pandemic influenza vaccination program.
### What (other) tasks could the dataset be used for? Are there obvious tasks for which it should not be used?
	An obvious task that VAERS should not be used for is for the purpose of providing medical advice. This dataset can be used to do a retrospective analysis or an investigative analysis of different variables that can infer from year to year what has been the trend in vaccination formulation and if there is any correlation between the side effects. 

From the website itself: “VAERS is not designed to determine if a vaccine caused a health problem but is especially useful for detecting unusual or unexpected patterns of adverse event reporting that might indicate a possible safety problem with a vaccine.”

### Has the dataset been used for any tasks already? If so, where are the results so others can compare (e.g., links to published papers)?
	It has been used in numerous academic publications that focus on vaccination safety and monitoring. This website links to academic papers by dataset year: https://www.cdc.gov/vaccinesafety/ensuringsafety/monitoring/vaers/publications.html

### Who funded the creation of the dataset? If there is an associated grant, provide the grant number.
	VAERS was created and is being co-managed by the Centers for Disease Control and Prevention along with the U.S. Food and Drug Administration, agencies of the U.S. Department of Health and Human Services. 

## __*Dataset Composition:*__

### What are the instances? (that is, examples; e.g., documents, images, people, countries) Are there multiple types of instances? (e.g., movies, users, ratings; people, interactions between them; nodes, edges) 
	The instances are the people who have reported events through the VAERS system. Anyone can report to the database. Health professionals, patients, parents, caregivers, etc. Their names are de-identified once reported. 

### Are relationships between instances made explicit in the data (e.g., social network links, user/movie ratings, etc.)? 
	There are no known relationships between instances besides the fact they are all individuals who willingly reported after a vaccination event. 

### How many instances of each type are there? 
What are the instances? (that is, examples; e.g., documents, images, people, countries) Are there multiple types of instances? (e.g., movies, users, ratings; people, interactions between them; nodes, edges) 
	There are 44,504 instances. Unknown currently if an individual reported more than once as each reported event is given its own unique VAERS ID number. 

### What data does each insistence consist of? 
	Consists of raw data collected from fields that instances would input once a event occurred. Data includes demographic information, type of vaccination, classification, and narratives of adverse symptoms and the resulting measures taken. 

### Is everything included or does the data rely on external resources? 
	Since VAERS is a passive reporting system, it means that it relies on individuals to send in reports of their experiences to the CDC and FDA. Therefore, this data does not collect all vaccination events across the U.S., adverse event or not.
