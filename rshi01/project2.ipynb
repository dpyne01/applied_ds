To take the total number of registered cars of related types into consideration, the dataset that recorded all vehicle registration information in New York State need to be merged. However, the dataset contains more than ten million rows and is too big to be merged with. Thus the downloadNYCRMV function was used to extract data of related car types from the origin dataset and record the number of registration cars for all related types. This function can return a dictionary that has the total number of registered cars of different types. There was a "typeConvert" dictionary that can help to convert different car type names to its uniform name. 

```python
import collections
import urllib.parse
def downloadNYCRMV():
    url = 'https://data.ny.gov/resource/w4pv-hbkt.csv?'
    typeUrl = 'https://data.ny.gov/resource/w4pv-hbkt.csv?city=NEW YORK       &body_type=%s&$limit=10000000'
    classUrl = 'https://data.ny.gov/resource/w4pv-hbkt.csv?city=NEW YORK       &registration_class=%s&$limit=10000000'
    types = ['SEDN', 'SUBN', '4DSD', '2DSD', 'VAN ']
    classes = ['MOT', 'PAS']
    typeConvert = {'SEDN': 'sedan', 'SUBN': 'suv', '4DSD': 'sedan', '2DSD': 'sedan', 'TR/C': 'truck', 'SWT ': 'truck', 'PICK': 'pick-up', 'VAN ': 'van', 'MOT': 'motorcycle', 'PAS': 'passenger vehicle'}
    typeMap = collections.defaultdict(int)
    for t in types:
        query = {'city': 'NEW YORK       ', '$limit': '10000000'}
        print('Type: %s' % t)
        query['body_type'] = t
        rUrl = url + urllib.parse.urlencode(query)
        print(rUrl)
        data = pd.read_csv(rUrl)
        typeMap[typeConvert[t]] += len(data)
    for c in classes:
        query = {'city': 'NEW YORK       ', '$limit': '10000000'}
        print('Type: %s' % c)
        query['registration_class'] = c
        rUrl = url + urllib.parse.urlencode(query)
        print(rUrl)
        data = pd.read_csv(rUrl)
        typeMap[typeConvert[c]] += len(data)
    return typeMap
typeMap = downloadNYCRMV()
```

Below showed the return dictionary of the downloadNYCRMV function. The keys were car types and the values were the number of cars for each type.

### 6. Machine Learning

Accidents that lead to people injured/killed is the type of accident people want to avoid the most. A lot of factors can relate with this problem. A collision can be defined by time, location and involved vehicle type. A machine learning model can be trained By taking these factors into consideration. In the origin data, the vehicle type is text data. A dictionary that contains represented numbers for vehicle types was used for converting text data to numerical data. Since this project planed to predict total number of persons injured/killed, the 'NUMBER OF PERSONS INJURED' column needed to be combined with 'NUMBER OF PERSONS KILLED' column. 

```python
total = data2018['NUMBER OF PERSONS KILLED']+data2018['NUMBER OF PERSONS INJURED']
data2018['fatalCount'] = total
dataNew = data2018[["TIME","ZIP CODE", "VEHICLE TYPE CODE 1", "fatalCount"]]
typeCode = {"suv":1, "sedan":2, "passenger vehicle":3, "taxi":4, "pick-up":5, "truck":6, "bus":7, 
        "bike":8, "van":9,"motorcycle":10,"tractor":11, "convertible":12,"ambulance":13,"garbage truck":14}
dropIndex = []
for i in range(len(dataNew.index)):
        if pd.isna(dataNew['VEHICLE TYPE CODE 1'][i]):
            continue
        typeName = dataNew['VEHICLE TYPE CODE 1'][i]
        if typeName in typeCode:
            dataNew.at[i, 'VEHICLE TYPE CODE 1'] = typeCode[typeName]
        else:
            dropIndex.append(dataNew.index[i])
dataNew = dataNew.drop(dropIndex)
dataNew = dataNew.dropna()
dataNew['fatalCount'] = dataNew['fatalCount'].astype(int)
```

After processing the initial data, the data was splited to train set and test set with test size equaled to 0.33. The variable that was planed to predict was the number of persons injured/killed. The observed data included vehicle type, zip code which showed the location information and time.  

```python
from sklearn.model_selection import train_test_split
x = dataNew.drop('fatalCount', axis=1)
X_train, X_test, y_train, y_test = train_test_split(x, dataNew.fatalCount, test_size=0.33)
```

Since the number of persons injured/killed can not have decimal, the predicted results were rounded to nearest interger. The machine learning model random forest was used for prediction. The model was first trained by train set then used to predict using test set. The predict result was compared with the label in the test set. The train accuracy and test accuracy was calculated seperately. The result showed that the train accuracy was 0.8 and the test accuracy was 0.75.

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import accuracy_score
import math
def roundL(l):
    return [round(x) for x in l]
def print_score(m):
    train_res = roundL(m.predict(X_train))
    test_res = roundL(m.predict(X_test))
    res = ["train accuracy:", accuracy_score(y_train, train_res), "test accuracy:", accuracy_score(y_test,test_res)]
    print(res)
m = RandomForestRegressor(n_estimators=100)
m.fit(X_train, y_train)
```
