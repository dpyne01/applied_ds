## 6. Machine Learning from Restaurants Reviews

```python
#Shape of the dataset
print(ta.shape)
```

Here we have 125527 instances and 7 attributes. I decided to use rating and number of views to forcast the price level of a restaurant here.

```python
median_rating = ta['rating'].median()
print("Median rating is {}".format(median_rating))
```

```python
# Use median rating number to replace NaN/missing value
ta['rating'].fillna(median_rating, inplace = True)
ta.rating.head()
```

```python
median_view = ta['no_views'].median()
print("Median number of views is {}".format(median_view))
```

```python
# Use median view count number to replace NaN/missing value
ta['no_views'].fillna(median_view, inplace = True)
ta.no_views.head()
```

```python
#Set up data inputs for machine learning
data_inputs = ta[["rating", "no_views"]]
data_inputs.head()
```

```python
#Set up expected output
ta['price_range'].fillna('Medium')
expected_output = ta[["price_range"]]
expected_output["price_range"].replace("High", 3, inplace = True)
expected_output["price_range"].replace("Medium", 2, inplace = True)
expected_output["price_range"].replace("Low", 1, inplace = True)
expected_output["price_range"].replace("Unknown", 0, inplace = True)
expected_output.head()
```

For me, number is better to understand for machine learning, so I chose to replace the strings with numbers.

```python
inputs_train, inputs_test, expected_output_train, expected_output_test = train_test_split (data_inputs, expected_output, test_size = 0.33, random_state = 42)
```

test_size=0.33 means 33% of the sample is to be used for testing, the other for training. random_state is used to initialise the inbuilt randomiser, so we get the same result from the randomiser each time.

```python
print(inputs_train.head())
print(expected_output_train.head())
```

```python
rf = RandomForestClassifier (n_estimators=100)
rf.fit(inputs_train, expected_output_train)
```

```python
#See how accurate my algorithm is
accuracy = rf.score(inputs_test, expected_output_test)
print("Accuracy = {}%".format(accuracy * 100))
```

This machine learning model is acceptable since the accuracy is around 70%.
